# Binary Cls

```python
from keras import Sequential
from keras.layers import Dense, Input
from sklearn.model_selection import train_test_split
```

```python
import kagglehub
```

## Data

```python
import pandas as pd
df=pd.read_csv("titanic_data.csv")
```

```python
df
```

```python
print(df.isna().sum())
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
scaler = StandardScaler()
df[['pclass', 'sibsp', 'parch', 'fare']] = scaler.fit_transform(df[['pclass', 'sibsp', 'parch', 'fare']])
```

```python
le = LabelEncoder()
for col in ['sex', 'embarked', 'alone']:
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
y=df['survived']
x= df[['pclass', 'sex',	'sibsp',	'parch',	'fare',	'embarked',	'alone','survived']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Model

- input layer neuorans ( 8 as 8 features )
---
- output layer activation function ( sigmoid )
- output layer Loss Function ( binary cross entropy )
- output layer neourans ( one )
---
- hidden layers number ( 2)
- activation on the hidden layers   ( relu )
- nuorans on hidden layers  ( 9neurons(6+3) )

```python

model = Sequential()

model.add(Dense(6, input_shape=(8,), activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.summary()

```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=32)
```

```python
df = pd.read_csv( "/kaggle/input/nursery/nursery.csv")
```

```python
label_encoders = {}
for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
X = df.drop("final evaluation", axis=1)
y = df["final evaluation"]
```

```python
y
```

```python
X.shape
```

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
len(set(y)) #num of classes in target
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, 5)
y_test= to_categorical(y_test, 5)
```

- input layer neuorans ( 8 as 8 features )
- output layer activation function ( softmax )
- output layer Loss Function ( categorical cross entropy )
- output layer neourans ( 5 )
- hidden layers number ( 2)
- activation on the hidden layers ( relu )
- nuorans on hidden layers ( 9neurons(6+3) )

```python
model = Sequential()

model.add(Dense(6, input_shape=(8,), activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(5, activation='softmax'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, validation_data=(X_test, y_test),epochs=100, batch_size=32)
```
